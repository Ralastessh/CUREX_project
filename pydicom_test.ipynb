```python
import numpy as np
import cv2, pydicom
from pydicom import dcmread
import matplotlib.pyplot as plt
from pydicom.pixel_data_handlers.util import apply_modality_lut, apply_voi_lut

window_center = 500
window_width = 100

dicom_path = './DATA/5thPost2/POST_1.CT.0004.0001.2022.04.26.18.00.48.734375.18710628.IMA'
image = pydicom.dcmread(dicom_path)

plt.subplot(1, 2, 1)
plt.title('DICOM -> Array(normal)')
plt.imshow(image.pixel_array, cmap = plt.cm.bone)
#[221213_1558]NOTED 원본 이미지 출력 -> 기존 WL, WW 반영 안 됨...?

image.WindowCenter = window_center
image.WindowWidth = window_width
image1 = apply_modality_lut(image.pixel_array, image)
image2 = apply_voi_lut(image1, image)

plt.subplot(1, 2, 2)
plt.title('apply_voi_lut')
plt.imshow(image2, cmap = plt.cm.bone)
```
